#### **Part 4 - Define your variable to extarct**

- API key: https://www.census.gov/data/developers/data-sets.html
- Extract data using API examples: https://api.census.gov/data/2022/acs/acs5/examples.html
- variables lookup: https://api.census.gov/data/2022/acs/acs5/variables.html

```python
#define the variables to be used in the API call

variables = [
    'NAME',     # Name of the area
    'B19013_001E',  # Median Household Income
    'B19013A_001E',  # Median Household Income (White Alone)
    'B19013B_001E'  # Median Household Income (Black or African American Alone)
     ] # Asian alone population
```

```python
# define years from 2003 t0 2021 to extract data from the API
years = list(range(2009, 2022))
```

#### **Part 5 - Extract Data**

```python
#loop over years and get data for each year and concatenate the data into a single dataframe
df = pd.DataFrame()
for year in years:
    print(year)
    data = c.acs.state(('NAME',) + tuple(variables), Census.ALL, year=year)
    df = pd.concat([df, pd.DataFrame(data)])
```

```python
# dataframe and its shape
df
```

```python
# average median household income for all states
avg_df = pd.DataFrame()
avg_df['Avg medium income'] = df.groupby('NAME')['B19013_001E'].mean()
avg_df['Avg medium income white'] = df.groupby('NAME')['B19013A_001E'].mean()
avg_df['Avg medium income black'] = df.groupby('NAME')['B19013B_001E'].mean()
```

```python
# dataframe and its shape
avg_df
```

#### **Part 6 - Save the dataframe**

```python
# save the dataframe to a csv file for future use
avg_df.to_csv('avg_median_income.csv')
```
