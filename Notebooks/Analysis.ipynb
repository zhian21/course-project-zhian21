```python
# Identify the last steps of each iteration
last_steps = df.groupby(['RunId', 'iteration']).agg({'Step': 'max'}).reset_index()

# Merge this back to the original DataFrame to get the rows corresponding to the last steps of each iteration
last_steps = pd.merge(df, last_steps, how='inner', on=['RunId', 'iteration', 'Step'])

# Calculate mean and standard deviation for time investment and child outcome scores by education level
summary_stats = last_steps.groupby('primary_edu_level').agg({
    'Average Time Investment (High)': ['mean', 'std'],
    'Average Time Investment (Medium)': ['mean', 'std'],
    'Average Time Investment (Low)': ['mean', 'std'],
    'Average Child Outcome Score (High)': ['mean', 'std'],
    'Average Child Outcome Score (Medium)': ['mean', 'std'],
    'Average Child Outcome Score (Low)': ['mean', 'std']
}).reset_index()

# Rename columns for better readability
summary_stats.columns = [
    'Education Level',
    'Mean Time Investment (High)', 'Std Time Investment (High)',
    'Mean Time Investment (Medium)', 'Std Time Investment (Medium)',
    'Mean Time Investment (Low)', 'Std Time Investment (Low)',
    'Mean Child Outcome Score (High)', 'Std Child Outcome Score (High)',
    'Mean Child Outcome Score (Medium)', 'Std Child Outcome Score (Medium)',
    'Mean Child Outcome Score (Low)', 'Std Child Outcome Score (Low)'
]

# Display the summary statistics table
print(summary_stats)
```

```python
# Identify the last steps of each iteration
last_steps = df.groupby(['RunId', 'iteration']).agg({'Step': 'max'}).reset_index()

# Merge this back to the original DataFrame to get the rows corresponding to the last steps of each iteration
last_steps = pd.merge(df, last_steps, how='inner', on=['RunId', 'iteration', 'Step'])

# Calculate mean and standard deviation for time investment and child outcome scores by education level
summary_stats = last_steps.groupby('primary_edu_level').agg({
    'Average Time Investment (High)': ['mean', 'std'],
    'Average Time Investment (Medium)': ['mean', 'std'],
    'Average Time Investment (Low)': ['mean', 'std'],
    'Average Child Outcome Score (High)': ['mean', 'std'],
    'Average Child Outcome Score (Medium)': ['mean', 'std'],
    'Average Child Outcome Score (Low)': ['mean', 'std']
}).reset_index()

# Flatten the multi-level columns
summary_stats.columns = [
    'Education Level',
    'Mean Time Investment (High)', 'Std Time Investment (High)',
    'Mean Time Investment (Medium)', 'Std Time Investment (Medium)',
    'Mean Time Investment (Low)', 'Std Time Investment (Low)',
    'Mean Child Outcome Score (High)', 'Std Child Outcome Score (High)',
    'Mean Child Outcome Score (Medium)', 'Std Child Outcome Score (Medium)',
    'Mean Child Outcome Score (Low)', 'Std Child Outcome Score (Low)'
]

# Format the results
def format_stats(mean, std):
    return f"{mean:.2f} ({std:.2f})"

formatted_summary = summary_stats.apply(lambda row: [
    format_stats(row['Mean Time Investment (High)'], row['Std Time Investment (High)']),
    format_stats(row['Mean Time Investment (Medium)'], row['Std Time Investment (Medium)']),
    format_stats(row['Mean Time Investment (Low)'], row['Std Time Investment (Low)']),
    format_stats(row['Mean Child Outcome Score (High)'], row['Std Child Outcome Score (High)']),
    format_stats(row['Mean Child Outcome Score (Medium)'], row['Std Child Outcome Score (Medium)']),
    format_stats(row['Mean Child Outcome Score (Low)'], row['Std Child Outcome Score (Low)'])
], axis=1, result_type='expand')

# Set column names for the formatted summary
formatted_summary.columns = [
    'Time Investment (High)', 'Time Investment (Medium)', 'Time Investment (Low)',
    'Child Outcome Score (High)', 'Child Outcome Score (Medium)', 'Child Outcome Score (Low)'
]

# Display the formatted summary statistics table
display(formatted_summary)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Identify the last steps of each iteration
last_steps = df.groupby(['RunId', 'iteration']).agg({'Step': 'max'}).reset_index()

# Merge this back to the original DataFrame to get the rows corresponding to the last steps of each iteration
last_steps = pd.merge(df, last_steps, how='inner', on=['RunId', 'iteration', 'Step'])

# Prepare data for boxplot
boxplot_data = last_steps.melt(id_vars=['primary_edu_level'], value_vars=[
    'Average Time Investment (High)',
    'Average Time Investment (Medium)',
    'Average Time Investment (Low)'
], var_name='Education Level', value_name='Time Investment')

# Map the education level names to a more readable format
education_level_map = {
    'Average Time Investment (High)': 'High',
    'Average Time Investment (Medium)': 'Medium',
    'Average Time Investment (Low)': 'Low'
}
boxplot_data['Education Level'] = boxplot_data['Education Level'].map(education_level_map)

# Define the colors for each education level
palette = {
    'High': '#2F4F4F',     # darkslategray
    'Medium': '#5F9EA0',   # cadetblue
    'Low': '#AFEEEE'       # paleturquoise
}

# Create the boxplot without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education Level', y='Time Investment', data=boxplot_data, showfliers=False, palette=palette)
plt.title(' Time Investment for Different Education Levels (Without Outliers)')
plt.xlabel('Education Level')
plt.ylabel('Time Investment')
plt.show()
```

```python
# Filter data for individual learning ratios
df_03 = df[df['high_individual_learning_ratio'] == 0.3]
df_07 = df[df['high_individual_learning_ratio'] == 0.7]

# Prepare data for plotting
plot_data_03 = df_03.melt(id_vars=['Step'], value_vars=[
    'Average Time Investment (High)',
    'Average Time Investment (Medium)',
    'Average Time Investment (Low)'
], var_name='Education Level', value_name='Time Investment')

plot_data_07 = df_07.melt(id_vars=['Step'], value_vars=[
    'Average Time Investment (High)',
    'Average Time Investment (Medium)',
    'Average Time Investment (Low)'
], var_name='Education Level', value_name='Time Investment')

# Map the education level names to a more readable format
education_level_map = {
    'Average Time Investment (High)': 'High',
    'Average Time Investment (Medium)': 'Medium',
    'Average Time Investment (Low)': 'Low'
}
plot_data_03['Education Level'] = plot_data_03['Education Level'].map(education_level_map)
plot_data_07['Education Level'] = plot_data_07['Education Level'].map(education_level_map)

# Define the colors for each education level
palette = {
    'High': '#2F4F4F',     # darkslategray
    'Medium': '#5F9EA0',   # cadetblue
    'Low': '#AFEEEE'       # paleturquoise
}

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# Plot for individual learning ratio = 0.3
sns.lineplot(ax=axes[0], data=plot_data_03, x='Step', y='Time Investment', hue='Education Level', palette=palette)
axes[0].set_title('Individual Learning Ratio = 0.3')
axes[0].set_xlabel('Step')
axes[0].set_ylabel('Time Investment')
axes[0].legend(title='Education Level')

# Plot for individual learning ratio = 0.7
sns.lineplot(ax=axes[1], data=plot_data_07, x='Step', y='Time Investment', hue='Education Level', palette=palette)
axes[1].set_title('Individual Learning Ratio = 0.7')
axes[1].set_xlabel('Step')
axes[1].set_ylabel('Time Investment')
axes[1].legend(title='Education Level')

plt.suptitle('Impact of Individual Learning Ratio on Time Investment Over Time')
plt.tight_layout()
plt.show()
```
